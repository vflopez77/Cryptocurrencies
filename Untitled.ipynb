```python
# Import Libraries
import pandas as pd
```

```python
# Load file to DataFrame
file_path = "Resources/iris.csv"
iris_df = pd.read_csv(file_path)
iris_df
```

```python
# Drop class column
new_iris_df = iris_df.drop(['class'], axis=1)
new_iris_df
```

```python
# Reorder Columns
new_iris_df = new_iris_df[['sepal_length', 'petal_length', 'sepal_width', 'petal_width']]
new_iris_df
```

```python
# Create csv file from cleaned data
output_file_path = "Resources/new_iris_data.csv"
new_iris_df.to_csv(output_file_path, index=False)
```

```python
# Data Loading
file_path = "Resources/shopping_data.csv"
df_shopping = pd.read_csv(file_path, encoding="ISO-8859-1")
df_shopping
```

```python
# Columns
df_shopping.columns
```

```python
# List DataFrame Types
df_shopping.dtypes
```

```python
# Find null values
for column in df_shopping.columns:
    print(f"Column {column} has {df_shopping[column].isnull().sum()} null values")
```

```python
# Drop null values
df_shopping = df_shopping.dropna()
df_shopping
```

```python
# Find Duplicate Entries
print(f"Duplicate entries: {df_shopping.duplicated().sum()}")
```

```python
# Remove the CustomerID Column
df_shopping.drop(columns=["CustomerID"], inplace=True)
df_shopping
```

```python
# Transform String Column
def change_string(member):
    if member == "Yes":
        return 1
    else:
        return 0
    
df_shopping["Card Member"] = df_shopping["Card Member"].apply(change_string)
df_shopping
```

```python
df_shopping.dtypes
```

```python
# Transform annual income to a similar scale
df_shopping["Annual Income"] = df_shopping["Annual Income"] / 1000
df_shopping
```

```python
df_shopping.rename(columns={"Card Member" : "Card_Member", "Annual Income" : "Annual_Income", "Spending Score (1-100)" : "Spending_Score"}, inplace=True)
df_shopping
```

```python
# Saving cleaned data
file_path = "Resources/shopping_data_cleaned.csv"
df_shopping.to_csv(file_path, index=False)
```

```python
# Saving cleaned data
file_path = "Resources/shopping_data_cleaned.tsv"
df_shopping.to_csv(file_path, sep="\t", index=False)
```
